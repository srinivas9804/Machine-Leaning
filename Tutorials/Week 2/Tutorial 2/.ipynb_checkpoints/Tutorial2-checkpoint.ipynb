### 1. Build document vectors
Create the vocabulary from the training data.  Then build the document vectors for the training and testing sets.  You can decide how many words you want in the vocabulary.

```python
### INSERT YOUR CODE HERE
cntvect = feature_extraction.text.CountVectorizer(stop_words='english', max_features=150)
trainX = cntvect.fit_transform(newsgroups_train.data)
testX  = cntvect.transform(newsgroups_test.data)
print(cntvect.vocabulary_)
```

```python
def showVocab(vocab, counts=None):
    "print out the vocabulary. if counts specified, then only print the words w/ non-zero entries"
    allwords = list(vocab.keys())
    allwords.sort()  # sort vocabulary by index
    wordlist = []
    for word in allwords:
        ind = vocab[word]
        if (counts is None):
            mystr = "{:3d}. {}".format(ind, word)
        elif (counts[0,ind]>0):
            mystr = "{:3d}. ({:0.4f}) {}".format(ind, counts[0,ind], word)
        else:
            continue  # skip it
        wordlist.append(mystr)

    # print 2 columns
    it = iter(wordlist)
    try:
        for i in it:
            print('{:<30}{}'.format(i, next(it)))
    except StopIteration:
        pass
    finally:
        del it

            
# show the vocabulary
showVocab(cntvect.vocabulary_)
```

```python
showVocab(cntvect.vocabulary_, trainX[0])
print("---")
print(newsgroups_train.data[0])
```

### 2. Bernoulli Naive Bayes 
Learn a Bernoulli Naive Bayes model from the training set.  What is the prediction accuracy on the test set?  Try different parameters (alpha, max_features, etc) to get the best performance.

```python
### INSERT YOUR CODE HERE
bmodel = naive_bayes.BernoulliNB(alpha=0.1)
bmodel.fit(trainX, newsgroups_train.target)
```

What are the most informative words for each category?  Run the below code.

Note: `model.coef_[i]` will index the scores for the i-th class

```python
# get the word names
fnames = asarray(cntvect.get_feature_names())
for i,c in enumerate(newsgroups_train.target_names):
    tmp = argsort(bmodel.coef_[i])[-10:]
    print("class", c)
    for t in tmp:
        print("    {:9s} ({:.5f})".format(fnames[t], bmodel.coef_[i][t]))
```

```python
predY = bmodel.predict(testX)
print("predictions: ", predY)
print("actual:      ", newsgroups_train.target)
acc = metrics.accuracy_score(newsgroups_test.target, predY)
print(acc)
```

### 3. Multinomial Naive Bayes model
Now learn a multinomial Naive Bayes model using the TF-IDF representation for the documents.  Again try different parameter values to improve the test accuracy.

```python
### INSERT YOUR CODE HERE
tf_trans = feature_extraction.text.TfidfTransformer(use_idf=True, norm='l1')
trainXtf = tf_trans.fit_transform(trainX)
testXtf = tf_trans.transform(testX)
mmodel_tf
```

```python
showVocab(cntvect.vocabulary_, trainXtf[0])
```

```python
def trainMultinomialBayes(val=0.2):
    global mmodel_tf
    mmodel_tf = naive_bayes.MultinomialNB(alpha=val)
    mmodel_tf.fit(trainXtf, newsgroups_train.target)
```

What are the most informative features for Multinomial model? Run the below code.

```python
# get the word names
trainMultinomialBayes()
fnames = asarray(cntvect.get_feature_names())
for i,c in enumerate(newsgroups_train.target_names):
    tmp = argsort(mmodel_tf.coef_[i])[-10:]
    print("class", c)
    for t in tmp:
        print("    {:9s} ({:.5f})".format(fnames[t], mmodel_tf.coef_[i][t]))
```

```python
def testMultinomialBayes():
    global predYtf
    predYtf = mmodel_tf.predict(testXtf)
    #print("prediction: ", predYtf)
    #print("actual:     ", newsgroups_test.target)
    acc = metrics.accuracy_score(newsgroups_test.target, predYtf)
    #print(acc)
    return acc
```

How do the most informative words differ between the TF-IDF multinomial model and the Bernoulli model?

- The TF-IDF model ignores words which appear in multiple different classes and tries to give importance to words which appear only in that specific class. Take the class comp.graphics for example, in the bernoulli model, words like 'just' and 'like' are common but these are common words which will be in different classes. In the TF-IDF model these words are replaced with more specific words of that class like 'file' and 'image'. 

Finally, look at a few of the misclassified documents/. 

```python
### INSERT YOUR CODE HERE ###
testMultinomialBayes()
ctr = 0 
#having a look at the first 10 misclassified elements
for i in range(len(predYtf)):
    if ctr > 10:
        break
    if predYtf[i] != newsgroups_test.target[i]:
        print("actual value : ",newsgroups_test.target[i],"predicted value : ",predYtf[i])
        ctr = ctr + 1
```

Can you get any intuition or reason why they were misclassified?

- Classes 0 and 3 are generally misclassified as the other. This is because atheism and religion are two topics which are inter-linked. The text will contain similar phrases and will talk about the other topic, hence it is more difficult to classify them.

### 4. Effect of smoothing
The smoothing (regularization) parameter has a big effect on the performance.  Using the Multinomial TF-IDF models, make a plot of accuracy versus different values of alpha. For each alpha, you need to train a new model. Which alpha value yields the best result?

```python
### INSERT YOUR CODE HERE
alphavalues = logspace(-5,0,50)
max_acc = 0
max_alpha = 0
accuracy = []
for i in alphavalues:
    trainMultinomialBayes(i)
    temp = testMultinomialBayes()
    accuracy.append(temp)
    if temp > max_acc:
        max_acc = temp
        max_alpha = i
plt.figure()
plt.semilogx(alphavalues, accuracy)
plt.semilogx(max_alpha, max_acc, 'kx')
plt.xlabel('alpha'); plt.ylabel('accuracy')     
plt.grid(True)  
plt.title('accuracy versus alpha');
```

```python
print('Max accuracy of classifer is :',max_acc," when alpha =",max_alpha)
```

### 5. Effect of vocabulary size
The vocabulary size also affects the accuracy.  Make another plot of accuracy versus vocabulary size.  Which vocabulary size yields the best result?

```python
### INSERT YOUR CODE HERE
```

```python
vocabsize = linspace(100,26000,20)
max_acc = 0
max_vocab = 0
accuracy = []
for i in vocabsize:
    cntvect = feature_extraction.text.CountVectorizer(stop_words='english', max_features=int(i))
    trainX = cntvect.fit_transform(newsgroups_train.data)
    testX  = cntvect.transform(newsgroups_test.data)
    tf_trans = feature_extraction.text.TfidfTransformer(use_idf=True, norm='l1')
    trainXtf = tf_trans.fit_transform(trainX)
    testXtf = tf_trans.transform(testX)
    trainMultinomialBayes(0.001)
    temp = testMultinomialBayes()
    accuracy.append(temp)
    if temp > max_acc:
        max_acc = temp
        max_vocab = i
plt.figure
plt.plot(vocabsize,accuracy)
plt.plot(max_vocab, max_acc, 'kx')
plt.grid(True)
plt.ylabel('accuracy')
plt.xlabel('vocabulary size')
plt.title('graph of vocabulary size vs accuracy')
plt.show()
```
